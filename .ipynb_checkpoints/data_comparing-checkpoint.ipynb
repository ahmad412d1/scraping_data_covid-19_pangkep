```python
#import library
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML
from pandas.io.formats.style import Styler
```

```python
#read the data
data_6 = pd.read_csv("data/6 mei 2021.csv")
data_7 = pd.read_csv("data/7 mei 2021.csv")
data_7
```

```python
#showing the increment of covid-19 cases
data_covid_pangkep = data_7.iloc[11:12].sum().iloc[2:14] -  data_6.iloc[11:12].sum().iloc[2:14]
data_covid_pangkep
```

```python
#make into data frame
df_covid_pangkep = data_covid_pangkep.to_frame()
df_covid_pangkep = df_covid_pangkep.rename(columns = {0:'Jumlah penambahan'})
df_covid_pangkep
df_covid_pangkep.style.set_caption("Data Pantauan COVID-19 di Kabupaten Pangkajene Dan Kepulauan")
```

```python
#update today
T_data = data_7.iloc[11:12].T
T_data = T_data.drop(axis=0, index=["No","Kabupaten/Kota"])
T_data= T_data.rename(columns = {10:'Jumlah'})
T_data.style.set_caption("Data Pantauan COVID-19 di Kabupaten Pangkajene Dan Kepulauan")
```
